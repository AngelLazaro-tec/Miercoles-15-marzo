```python
import numpy as np
import matplotlib.pyplot as plt

# paso 1. obtener los datos
cats = ['A', 'B', 'C', 'D', 'E']
conteos = np.array([45, 80, 60, 30, 95])

# paso 2. figura y el eje
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()

# paso 3. graficar
barras = ax.bar(cats, conteos, color="skyblue")

# paso 4 etiquetas
ax.set_title("barras: ventas por categoria")
ax.set_xlabel("categoria")
ax.set_ylabel("ventas")
ax.grid(True, axis='y', linestyle="--", alpha=0.4)

# paso 5. agregar valores encima de cada barra
for barra in barras:
    altura = barra.get_height()
    ax.text(
        barra.get_x() + barra.get_width()/2,
        altura,
        f'{int(altura)}',
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# paso 1. datos (ventas en miles de pesos)
productos = ['Laptop', 'Celular', 'Tablet', 'Audifonos', 'Smartwatch', 'Teclado']
ventas = np.array([125.5, 210.3, 98.7, 75.2, 150.8, 60.4])

# paso 2. ordenar de mayor a menor
orden = np.argsort(ventas)[::-1]
ventas_ord = ventas[orden]
productos_ord = np.array(productos)[orden]

# paso 3. figura y eje
fig = plt.figure(figsize=(7, 4))
ax = plt.gca()

# paso 4. barras continuas (sin separación)
barras = ax.bar(range(len(ventas_ord)), ventas_ord, width=1.0, color="steelblue")

# paso 5. etiquetas eje x
ax.set_xticks(range(len(productos_ord)))
ax.set_xticklabels(productos_ord, rotation=30)

# paso 6. etiquetas y grid
ax.set_title("ventas por producto (miles de pesos)")
ax.set_xlabel("producto")
ax.set_ylabel("ventas")
ax.grid(True, axis='y', linestyle="--", alpha=0.4)

# paso 7. valores encima
for i, v in enumerate(ventas_ord):
    ax.text(i, v, f'{v:.1f}', ha='center', va='bottom')

plt.show()
```
